This notebook takes the numveh (base year vehicle stock data) as an input and produces the totnumveh (

```python
import pandas as pd
```

```python
# make a df for totnumveh

totnumveh = pd.DataFrame(columns=['CountryID', 'TransTypeID', 'VehTypeID', 'Age', 'TotNumVeh', 'TotNumVeh-1', 'NumVehScrap'])

totnumveh
```

```python
# make a vehtype dictionary
vehtype_dict = {2: ['Motorcycles'], 3: ['Cars'], 4: ['Buses and minibuses'], 
               5: ['LCVs', 'HGVs']}
```

```python
# read aemda output data - vehicle stock
stock = pd.read_csv(f'./out_data/VehicleStock.csv')
stock
```

```python
base_year = 2015
row_count = 0

for transtype in stock.TransTypeID.unique():
    
    for vehtype in stock.VehTypeID.unique():
            
        if not stock[(stock.TransTypeID == transtype) & (stock.VehTypeID == vehtype)].empty:

            for age in stock[(stock.TransTypeID == transtype) & (stock.VehTypeID == vehtype)].Age.unique():

                # return total stock of this config + age
                total_stock = stock[(stock.TransTypeID == transtype) & (stock.VehTypeID == vehtype) & (stock.Age == age)].NumVeh.sum()

                totnumveh.at[row_count, 'CountryID'] = 9
                totnumveh.at[row_count, 'TransTypeID'] = transtype
                totnumveh.at[row_count, 'VehTypeID'] = vehtype
                totnumveh.at[row_count, 'Age'] = age
                totnumveh.at[row_count, 'TotNumVeh'] = total_stock

                row_count +=1

# just fill it with zero scrap and -1 = '' for now and see what happens
totnumveh['TotNumVeh-1'] = totnumveh['TotNumVeh']
totnumveh['NumVehScrap'] = 0

totnumveh
```

```python
totnumveh.to_csv('totnumveh.csv')
```
