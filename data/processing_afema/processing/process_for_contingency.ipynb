```python
# read technology data - check for presence of SHI technologies
technology = pd.read_csv(f'./out_data/Technology.csv')
technology
```

```python
# return techIDs for buses + trucks (vehtypeid = 4 and 5)
techs = vehstock[(vehstock.VehTypeID == 4) | (vehstock.VehTypeID == 5)].TechID.unique().tolist()

for techID in techs:
    
    # if the techID minus 9000 is also in TechID, this TechID is a SHI
    if techID - 9000 in vehstock.TechID.tolist():
        
        # return ages present for this TechID 
        ages = vehstock[vehstock.TechID==techID]['Age'].unique().tolist()
        
        for age in ages:
            
            # return number of vehicles for this shi technology at this age
            numveh = vehstock[(vehstock.TechID==techID) & (vehstock.Age==age)].NumVeh.item()
            
            # return index of secondhand technology at this age
            shi_ind = vehstock[(vehstock.TechID==techID) & (vehstock.Age==age)].index.item()
            
            # if age is in the new tech
            if age in vehstock[(vehstock.TechID==techID - 9000)].Age.tolist():
                # return index of new technology at this age to add to
                new_ind = vehstock[(vehstock.TechID==techID - 9000) & (vehstock.Age==age)].index.item()
            else:
                new_ind = vehstock[(vehstock.TechID==techID - 9000) & (vehstock.Age==max(vehstock[(vehstock.TechID==techID - 9000)].Age.tolist()))].index.item()
        
            # add to number to vehicles in the new tech
            vehstock.at[new_ind, 'NumVeh'] += numveh
            
            # take away from number of vehicles in the shi tech (replace with 0)
            vehstock.at[shi_ind, 'NumVeh'] = 0
        
vehstock_new = vehstock[vehstock.NumVeh>0]
vehstock_new
```
